```python
# Função para executar uma query
def executar_query(query):
    return pd.read_sql(query, conn)
```

```python
query = """
SELECT
    TABLE_SCHEMA,
    TABLE_NAME
        FROM INFORMATION_SCHEMA.TABLES
        WHERE TABLE_TYPE = 'BASE TABLE'
        ORDER BY TABLE_SCHEMA, TABLE_NAME
"""

df_tabelas = executar_query(query)
df_tabelas
```

Nome dos clientes

```python
query_cliente = """
SELECT nomeCliente as Clientes
FROM clientes
"""

df_clientes = executar_query(query_cliente)
df_clientes
```

Produtos com valor > 2000 e descrição preenchida

```python
query_produtos = """
SELECT nomeProduto as Produtos, precoUnitario as Valor, descricaoProduto as Descricao
FROM produtos
WHERE precoUnitario > 2000
AND descricaoProduto is not null
"""

df_produtos = executar_query(query_produtos)
df_produtos
```
