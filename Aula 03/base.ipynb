```python
def geradorDeDesconto(valor, desconto = 0.1, *args, **kwargs): #Args = lista sem nome, kwargs = lista com nomes
    """
        Função calcula o desconto por % inserido

        Valor: pode ser float ou int
        
        Desconto: precisa ser informadop em decimal
    """

    return valor * desconto

geradorDeDesconto(1500, 0.2)
```

```python
# Quantas contas estão vencidas?

from datetime import datetime

hoje = datetime.today()

# print(hoje)

contas_vencidas = []

for conta in contas:
    if conta['vencimento'] < hoje and conta['status'] == 'Em aberto' and conta['tipo'] == 'Pagar':
        contas_vencidas.append(conta)

print(len(contas_vencidas))
contas_vencidas
```

```python
# Total em contas vencidas?

total_vencido = 0

for conta in contas_vencidas:
    total_vencido += conta['valor']

round(total_vencido,2)
```
