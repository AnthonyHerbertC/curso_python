Importando bibliotecas

```python
import os
import smtplib
import shutil
import pyautogui
import pywhatkit as kit
from time import sleep
from datetime import datetime
from email.message import EmailMessage
import win32com.client as win32
import pythoncom
```

Organizando arquivos

```python
arquivo_excel = 'contas_processadas.xlsx'
pasta_base = 'base'
```

Verificando se o arquivo existe

```python
if not os.path.exists(arquivo_excel):
    raise FileNotFoundError("Arquivo não encontrado, verifique o nome do arquivo")

print("Arquivo encontrado")
```


Criando a pasta base dentro da pasta Aula 04

```python
if not os.path.exists(pasta_base):
    os.makedirs(pasta_base)
    print('Pasta criada com sucesso')

else:
    print("Pasta base já existe")
```

Movendo arquivos para dentro da pasta Base

```python
destino = os.path.join(pasta_base,os.path.basename(arquivo_excel))

if not os.path.exists(destino):
    shutil.move(arquivo_excel, destino)
    print('Arquivo moviddo com sucesso')

else:
    print('Arquivo já existe na pasta')
```

Criando corpo do Email

```python
hoje = datetime.now().strftime('%d/%m/%Y | %H:%M')

corpo_email = f"""
Bom dia,

Segue em anexo o relatorio de contas processadas
Data de atualização: {hoje}

Atenciosamente
"""
```

Configurando Outlook

```python
pythoncom.CoInitialize()
outlook = win32.Dispatch('Outlook.Application')

email = outlook.CreateItem(0)

email.To = 'anthony.souza@jadlog.com.br'
email.Subject = 'Relatorio de contas processadas'
email.Body = corpo_email

email.attachments.Add(
    r'C:\Users\anthony.souza\OneDrive - JADLOG LOGISTICA LTDA\Documentos\Repositorio Frotas\Curso-Python\Materiais aulas\Aula 04\base\contas_processadas.xlsx'
)

email.Send()

print('E-mail enviado com sucesso!')
```

Alerta WhatsApp

```python
telefone = '+5511988153947'

mensagem = "Relatorio enviado"

kit.sendwhatmsg_instantly(
    telefone,
    mensagem,
    wait_time=20,
    tab_close=False
)

sleep(2)

pyautogui.press('enter')
print('mensagem enviada com sucesso')
```
