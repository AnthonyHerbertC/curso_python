Importando biblioteca

```python
import pandas as pd
```

Criação Dataframe

```python
df_entregas = pd.DataFrame({
    'id_entrega': [1,2,3],              # Chave primaria
    'filial'    : ['SP', 'RJ', 'MG'],
    'status'    : ['Entregue', 'Atrasado', 'Entregue']
})

df_custos = pd.DataFrame({
    'id_custo'   : [1,2,3],              # Chave primaria
    'id_entrega' : [1,1,3],              # Chave estrangeira
    'custo'       : [1000.5, 15000.20, 1255.25]
})
```

Join

```python
# Inner Join

df_inner = pd.merge(df_entregas, df_custos, on='id_entrega', how='inner')
print(df_inner)
```

```python
# Left Join

df_left = pd.merge(df_entregas, df_custos, on='id_entrega', how='left')
print(df_left)
```

```python
# Right Join

df_right = pd.merge(df_entregas, df_custos, on='id_entrega', how='right')
print(df_right)
```

```python
# full outer Join

df_full = pd.merge(df_entregas, df_custos, on='id_entrega', how='outer')
print(df_full)
```

```python
df_entregas_idx = df_entregas.set_index('id_entrega')
df_custos_idx = df_custos.set_index('id_entrega')

df_join = df_entregas_idx.join(df_custos_idx, how='left')
df_join.reset_index()
```
