```python
produtos = [
    ['Feijão', 4.99, 10, 'Sr Joao'],
    ['Arroz', 15.89, 15, 'Namorado'],
    ['Margarina', 6.99, 19, 'Qually'],
    ['Maionese', 8.89, 2, 'Helmans'],
    ['Maionese', 12.89, 23, 'Heinz'],
    ['Ketchup', 5.50, 25, 'Quero'],
    ['Mostarda', 4.89, 12, 'Quero']
]

Colunas_produtos = ['Produto','Preço', 'Estoque','Marca']

df_produtos = pd.DataFrame(produtos, columns=Colunas_produtos)

df_produtos
```

```python
df_produtos[df_produtos['Preço']>10]

df_produtos.describe()

df_produtos['Preço'].max()

df_produtos.groupby('Produto')['Preço'].max()

df_produtos.groupby('Produto').agg(
    Maior_preço = ('Preço', 'max'),
    Maior_Estoque = ('Estoque', 'max'),
    total_preço = ('Preço', 'sum')
)

display(df_produtos)
```
