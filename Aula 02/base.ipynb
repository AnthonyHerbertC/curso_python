```python
# repetindo um texto 5 vezes com for

for contador in range(5):
    print("Batata")

# repetindo um texto 5 vezes com while

numero = 0
while numero < 5:
    print (f"Frita")
    numero += 1
```

<p style="font-size: 1.1rem; color:#9beb3b; text-align: center; font-family: monospace ">
    <strong></br></br></br></br>
        Validando um senha de brincadeira
    </strong>
</p>

```python
senha_cadastrada = input ("Senha: ")
senha_entrada = ""
contador = 0

while senha_cadastrada != senha_entrada:
    senha_entrada = input("Digite a senha")
    contador += 1
    if contador == 3 and senha_cadastrada != senha_entrada:
        print("Excedeu numeros de tentativas")
        break

if senha_cadastrada == senha_entrada:
    print("Acesso permitido")
```

        While com Break
    </strong>
</p>

```python
while True:
    menu = int(input("1. Cadastrar | 2. Vizualizar | 3. Sair | Opção: "))
    match menu:
        case 1:
            franquia = input("Franquia: ")
        case 2:
            print(franquia)
        case 3:
            break
        case _:
            print("opção invalida")
```

    </strong>
</p>

```python
# Entrada de valores

franquias = ["Matriz SP", "FL Rio de janeiro", "CO São Paulo"]
vl_recebido = []

for indice, nome in enumerate(franquias):
    valor = float(input(f"Digite o valor recebido de {nome}: "))
    vl_recebido.append(valor)

print (franquias)
print (vl_recebido)

for indice, nome in enumerate(franquias):
    for id, valor in enumerate(vl_recebido):
        if indice == id:
            print(f"Valor recebido da franquia {nome} é de R${valor}")

print("total arrecadado R$",sum(vl_recebido))
```

```python
# Cálculo

for indice, nome in enumerate(franquias):
    for id, valor in enumerate(vl_recebido):
        if indice == id:
            vl_final = valor*0.1
            print(f"Valor recebido da franquia {nome} é de R${valor}, 10% do valor é igual à R${vl_final}")
```

<p style="font-size: 1.1rem; color: #3bebda; text-align: center; font-family: monospace ">
    <strong></br></br></br></br>
        # Exemplo: Cadastrando e consumindo Franquias  
    </strong>
</p>

```python
franquias_cadastradas = []

for contador in range(2):
    
    franquia = dict(
        id       = int(input("Identificação: ")),
        cpf      = input("CPF: "),
        endereco = input("Endereço: ").title(),
        bairro   = input("Bairro: ").title(),
        cidade   = input("Cidade: ").title(),
        estado   = input("Estado: ").upper()
    )

    franquias_cadastradas.append(franquia)
```

```python
for id, franquia in enumerate(franquias_cadastradas):
    id += 1
    print(f"{id}ª Franquia"),
    print(f"ID = {franquia['id']}"),
    print(f"CPF = {franquia['cpf']}"),
    print(f"Endereço = {franquia['endereco']}"),
    print(f"Bairro = {franquia['bairro']}"),
    print(f"Cidade = {franquia['cidade']}"),
    print(f"Estado = {franquia['estado']}")
    print("")
```

```python
for id, franquia in enumerate(franquias_cadastradas):
    id += 1
    print(f"{id}ª Franquia:")
    for chave, valor in franquia.items():
        print(f"{chave:<8} → {valor}")
    print()
```
