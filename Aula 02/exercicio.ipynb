üí° <strong>Dica:</strong> utilize <strong>while True</strong> para manter o menu em execu√ß√£o,
<strong>if</strong> para valida√ß√µes e <strong>match / case</strong> para tratar as op√ß√µes do menu.

```python
# Dados de entrada
from IPython.display import clear_output

saldo = 0.00
extrato = []

while True:
    menu = int(input('1. Entrada | 2. Saida | 3. Sair | Op√ß√£o'))
    
    if menu < 3 and menu > 0:
        valor = float(input('Digite o valor: R$'))
    
    match menu:
        case 1:
            clear_output()
            for cont in range (1):
                historico = dict(
                    Entrada = valor
                )
                extrato.append(historico)
            saldo += valor
            print(f'Entrada realizada | R${valor}\nSaldo disponivel | R${saldo}')
        case 2:
            clear_output()
        #    valor = float(input('Digite o valor de Sa√≠da: R$'))
            if valor <= saldo:
                for cont in range (1):
                    historico = dict(
                        Saida = valor
                    )
                    extrato.append(historico)
                saldo -= valor
                print(f'Saida realizada | R${valor}\nSaldo disponivel | R${saldo}')
            else:
                print(f"Saldo insuficiente\nSaldo disponivel | R${saldo}")
        case 3:
            clear_output()
            print('Programa encerrado')
            break
        case _:
            clear_output()
            print('Op√ß√£o invalida, digite novamente')

print()
print (f"Saldo final: R${saldo}")
print()
print("Extrato:")
for transacao in extrato:
    for chave, valor in transacao.items():
        print(f'{chave:<9} ‚Üí {valor}')
```
