```python
# Entrada de dados
Nome_do_funcion√°rio = str(input("Digite o nome: ")).title()
Cargo = int(input("Cargo: 1. Analista Financeiro | 2. Assistente Financeiro | 3. Coordenador"))
Salario = float(input("Salario: "))
Anos_de_Experiencia = int(input("Quantos anos de experiencia? "))
```

```python
# Identificar o cargo escolhido

match Cargo:
    case 1:
        Cargo = "Analista Financeiro"
    case 2:
        Cargo = "Assistente Financeiro"
    case 3:
        Cargo = "Coordenador"
    case _:
        Cargo = "Cargo invalido, verifique novamente"

Cargo
```

```python
# Classifica√ß√£o do funcion√°rio por tempo de experi√™ncia

if Anos_de_Experiencia <= 2 and Anos_de_Experiencia >= 0:
    Senioridade = "J√∫nior"

elif Anos_de_Experiencia <= 5 and Anos_de_Experiencia >= 3:
    Senioridade= "Pleno"

elif Anos_de_Experiencia >= 6 and Anos_de_Experiencia <= 40:
    Senioridade = "S√™nior"

else:
    Senioridade = "Anos de experiencia invalido"

Senioridade
```

```python
# Aplica√ß√£o de l√≥gica de decis√£o 

if Cargo == "Analista Financeiro" and Senioridade != "J√∫nior":
    Permiss√£o = "Gerenciamento Liberado"

elif Cargo == "Asistente Financeiro" and Senioridade == "J√∫nior":
    Permiss√£o = "Apenas opera√ß√µes basicas"

elif Cargo == "Coordenador" and Senioridade == "S√™nior":
    Permiss√£o = "Acesso Total"

elif Senioridade == "Anos de experiencia invalido" or Cargo == "Cargo invalido, verifique novamente":
    Permiss√£o = "Verifique as respostas"

else:
    Permiss√£o = "Permiss√£o negada"
```
